**Problem 4 :**
<pre>
def longest_increasing_subsequence(nums):
    n = len(nums)
    lis = [1] * n
    for i in range(1, n):
        for j in range(0, i):
            if nums[i] > nums[j] and lis[i] < lis[j] + 1:
                lis[i] = lis[j] + 1
    return max(lis)
</pre>

**Solution**
<pre>
The time complexity of the longest_increasing_subsequence function is O(n^2),
where n is the length of the input list nums.   

This is because the function uses two nested loops to iterate over the list.
The outer loop iterates over each element in the list, and the inner loop
iterates over all the elements that come before the current element.  

For each element in the list, the inner loop performs a constant-time
operation to check if the current element is greater than the element
it is being compared to. If it is, then the function updates the length of
the longest increasing subsequence that ends at the current element.   

Since the inner loop iterates over all the elements that come before the
current element, the total time complexity of the function is O(n^2).
</pre>

**Problem 6 : Sum of Digits**

Write a recursive function to calculate the sum of digits of a given positive integer.   
sum_of_digits(123) -> 6

```python
def sum_of_digits(n):
  if n == 0:
    return 0
  else:
    return n % 10 + sum_of_digits(n // 10)

print(sum_of_digits(123))
```

**Problem 7: Fibonacci Series**

Write a recursive function to generate the first n numbers of the Fibonacci series.   
fibonacci_series(6) -> [0, 1, 1, 2, 3, 5]

```python
def fibonacci_series(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        fib_series = fibonacci_series(n-1)
        next_number = fib_series[-1] + fib_series[-2]
        fib_series.append(next_number)
        return fib_series

print(fibonacci_series(6))  # Output: [0, 1, 1, 2, 3, 5]
```

**Problem 8 : Subset Sum**   
Given a set of positive integers and a target sum, write a recursive function to determine if there exists a subset
of the integers that adds up to the target sum.   
subset_sum([3, 34, 4, 12, 5, 2], 9) -> True

```python
def subset_sum(numbers, target_sum):
  if target_sum == 0:
    return True
  if not numbers:
    return False

  head, tail = numbers[0], numbers[1:]
  return subset_sum(tail, target_sum) or subset_sum(tail, target_sum - head)

print(subset_sum([3, 34, 4, 12, 5, 2], 9))  # True
```

**Problem 9: Word Break**  
Given a non-empty string and a dictionary of words, write a recursive function to determine if the string can be
segmented into a space-separated sequence of dictionary words.  
word_break( leetcode , [ leet , code ]) -> True

```python
def wordBreak(wordList , word):
  if word == "":
    return True
  else :
    wordLen = len(word)
    for i in range(1, wordLen+1):
      if word[:i] in wordList and wordBreak(wordList, word[i:]):
        return True
    return False

print(wordBreak(["leet","code"], "leetcodes"))
print(wordBreak(["I","am", "sam", "sung", "love", "vishwa"], "lovevishwa"))
```
